## Indexing and slicing of Array in Numpy

```python
# sLicing 1 D array
str_a = np.array(list("sagarmatha"))
```

```python
len_a = len(str_a)
len_a
```

```python
str_a[0] # leftmost character
```

```python
str_a[-1]
```

```python
# first three characters from left to right
str_a[0:3]
```

```python
# first three characters from right to left
str_a[-1:-4:-1]
```

```python
# slicing in 2D Array
l=['a','e','i','o','u']
x=['m','n','o','p','q']
j=['d','x','b','c','p']
p=['k','l','e','y','t']
```

```python
st=np.array([l,x,j,p])
st
```

```python
np.size(st)
```

```python
np.shape(st)
```

```python
st
```

```python
st[0:2,0:2]
```

```python
st[0:4,0:2]
```
