```python
# Import Matplotlib and Seaborn 
import matplotlib.pyplot as plt  # import matplotlib
import seaborn as sns  # import seaborn 
```

```python
# laod data cars 
import pandas as pd 
data = pd.read_csv(r"D:\python project\cars_data_cleaned.csv")
```

```python
data.head(2)
```

```python
# Bar Plot
# Countplot 
# Pie Chart 
# Line Char 
# Scatterplot 
# Stripplot 
# Heatmap
# Box plot / Violing plot
# KDE Plot 
```

```python
# Barplot 
plt.title("Average Mileage of Each Fuel Type")
sns.barplot(x='Fuel', y='Mileage',hue='Transmission',palette='bright',  ci=None, data=data)
plt.xticks(rotation=90,fontsize=8)
plt.show()
```

```python
#  Count Plot , this is used to count the frequency of items 
plt.title("Counts  of Cars Sold per Loocation")
sns.countplot(x='Location',  palette='cool', hue='Transmission', data=data)
plt.xticks(rotation=90)
plt.show()
```

```python
# Pie Charts 
data['Price Type'].value_counts()
```

```python
data['Price Type'].value_counts().values
```

```python
data['Price Type'].value_counts().index
```

```python
plt.title("Market Share of Cars According to Price Range")
plt.pie(data['Price Type'].value_counts().values, labels = data['Price Type'].value_counts().index, autopct='%1.1f%%')
plt.show()
```

```python
# Line Plot 
data.head(2)
```

```python
plt.figure(figsize=(20,5))
plt.title("Line Chart Showing Average Mileage")
sns.lineplot(x='Brand', y='Mileage',ci=None,  marker='D',data=data ,label="Average Mileage")
sns.lineplot(x='Brand', y='Power', ci=None, marker='*', data=data, label="Average Power")
plt.xticks(rotation=90)
#plt.label()
plt.show()
```

```python
 # Scatterplot() 
sns.scatterplot(x='Power', y='Mileage', hue='Transmission',data=data)
```

```python
#  Stripplot 
plt.title("Mileage Disctibution of Cars According to Transmission in Each Location")
sns.stripplot(x='Location', y='Mileage', hue='Fuel', data=data)
plt.xticks(rotation=90)
plt.show()
```
