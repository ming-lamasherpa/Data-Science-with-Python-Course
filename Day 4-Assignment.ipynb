## Q35. Write a program that takes a sentence, counts the total number of vowels (a, e, i, o, u) in it using a loop and count() or by checking each character.

```python
# Using loop
sentence = input("Enter a sentence: ").lower()
vowels = "aeiou"
count = 0

for char in sentence:
    if char in vowels:
        count += 1

print("Total number of vowels:", count)


# Using count() method
sentence = input("Enter a sentence: ").lower()
vowel_count = sum(sentence.count(v) for v in "aeiou")

print("Total number of vowels:", vowel_count)
```

## Q37. Write a program to count the frequency of each unique word in the sentence:'the cat sat on the mat and the cat sat'. Use split() and a dictionary.

```python
sentence = "the cat sat on the mat and the cat sat"
words = sentence.split()

word_freq = {}

for word in words:
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1

print("Word frequency:")
for word, count in word_freq.items():
    print(word, ":", count)
```

## Q45. Create a dictionary :scores = {'Alice': 85, 'Bob': 92, 'Charlie': 78, 'David': 95}. Write code to:Find the student with the highest score,Find the student with the lowest score, Print all students who scored above 80.

```python
scores = {'Alice': 85, 'Bob': 92, 'Charlie': 78, 'David': 95}

# Highest score
highest = max(scores, key=scores.get)
print("Highest scorer:", highest)

# Lowest score
lowest = min(scores, key=scores.get)
print("Lowest scorer:", lowest)

# Students scoring above 80
for name, score in scores.items():
    if score > 80:
        print("Above 80:", name)
```
