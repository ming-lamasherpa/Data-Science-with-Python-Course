```python
#  Add a conditonal colum, engine__type 
# >=2000,"Very Powerful", >=1000,"Powerful", '>=500' Moderate, 'lesspowerful
```

```python
# Create a function 
def classify(value): 
    if value>=2000: 
        return  "Very Powerful"
    elif value >=1000: 
        return "Powerful"
    elif value >=500:
        return "Moderate" 
    else: 
        return "Less Powerful"
```

```python
data['Engine Type'] = data['Engine'].apply(classify)
```

```python
# Grouping the values and finding summary 
# find the max  mileage of each location 
data['Mileage'].max()
```
