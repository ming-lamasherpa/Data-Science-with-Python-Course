## Dynamic Argumnets

```python
def dy(*v):
    s=0
    for x in v:
        s=s+x
    return s
```

```python
dy(10,21,30)
```

```python
dy(10)
```

```python
dy(91,92,81,36,21)
```

```python
# Semi-dynamic
def admission(name, faculty, *optns):
    print("Name:",name)
    print("Faculty:",faculty)
    for o in optns:
        print(o)
    return
```

```python
admission("Mingma","CSIT","Dharan",9804563368)
```

## Looping

```python
for x in range(1,11):
    print(x)
```

```python
for x in range(1,11):
    print(x,end="-")
```

```python
for x in range(1,51,3):
    print(x,end=' ')
```

```python
for x in range(90,-1,-2):
    print(x, end=' ')
```

```python
for x in range(5, 51, 5):
    print(x, end=" ")
```

```python
lst = list("Ming lama channel")
lst
```

```python
for e in lst:
    print(e)
```

```python
for x in range(1,9):
    for y in range(1,x):
        print(y,end=' ')
    print('\n')
```

```python
# printnumbers from 0 to 20
lc=0
while lc <= 20 :
    print(lc,end=' ')
    lc+=1
```
