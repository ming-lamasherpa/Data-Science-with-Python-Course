```python
# Use `.apply()` to create a 'Profit_Category' column: 'Loss' if Profit < 0, 'Low' if < 50, 'High' otherwise.
df['Profit_Category']=df['Profit'].apply(lambda x:'Loss' if x<0 else ('Low' if x<50 else 'High'))
```

```python
# Identify the top 10% of orders by Sales using `.quantile()`.
threshold=df['Sales'].quantile(0.9); df[df['Sales']>=threshold]
```

```python
# Use `pd.get_dummies()` to one-hot encode the `Ship Mode` column.
pd.get_dummies(df,columns=['ShipMode'])
```

```python
# Calculate the cumulative sum of `Sales` sorted by `Order Date`.
df.sort_values('Order Date')['Sales'].cumsum()
```

```python
# Use `.pipe()` to chain two custom functions: one that adds 'Shipping_Days' and one that adds 'Profit_Margin'.
def add_shipping(df):
    df['Shipping_Days']=(pd.to_datetime(df['Ship Date'])-df['Order Date']).dt.days
    return df

def add_margin(df):
    df['Profit_Margin']=df['Profit']/df['Sales']
    return df

df.pipe(add_shipping).pipe(add_margin)
```

```python
# Detect outliers in `Sales` using the IQR method (Q1 - 1.5×IQR and Q3 + 1.5×IQR).
Q1=df['Sales'].quantile(0.25); Q3=df['Sales'].quantile(0.75); IQR=Q3-Q1; df[(df['Sales']<Q1-1.5*IQR)|(df['Sales']>Q3+1.5*IQR)]
```

```python
# Use `itertools.groupby` equivalent in pandas to find consecutive same-day orders.
df['Same_Day_Group']=(df['Order Date'].diff()!=pd.Timedelta(0)).cumsum()
```

```python
# Create a multi-level column DataFrame by grouping on Region and computing multiple aggregations on Sales and Profit.
df.groupby('Region').agg({'Sales':['sum','mean'],'Profit':['sum','mean']})
```
