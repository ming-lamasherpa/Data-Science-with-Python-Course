```python
# select all the data where mileage is >=30
con =data['Mileage']>=30 
```

```python
data[con]
```

```python
# filter data with multiple condition  (and)
# select all the data where mileage >=30 and Transmission ='Manual' 
con1 = data['Mileage']>=30 
con2 = data['Transmission']=='Manual'  
```

```python
test = data[con1 & con2 ]
test.head(2)
```

```python
data[con1 | con2]
```

```python
# using not conditio 
# select all the values except Transmission is Manual 
data[data['Transmission']!='Manual']
```
