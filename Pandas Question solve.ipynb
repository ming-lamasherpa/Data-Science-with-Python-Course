## Q2. Create a Series using range(1, 20, 2) and display only the values at index positions 2, 4, and 6.

```python
import pandas as pd

s = pd.Series(range(1,20,2))
print(s[[2,4,6]])
```

## Q3. Create a DataFrame from the list [10, 20, 30, 40, 50] and rename its default column to "values".

```python
import pandas as pd

data = [10,20,30,40,50]
df = pd.DataFrame(data) #pd.DataFrame(data) → creates a table (DataFrame)
print(df)
```

```python
df.columns = ["values"] #Default column name is 0
print(df) #df.columns = ["values"] → changes column name
```

## Q4. Create a DataFrame from a dictionary with columns "Name", "Age", and "City" for at least 5 people. Display it.

## Q7. Create a DataFrame with 3 columns and 4 rows using a nested list. Assign custom column names.
