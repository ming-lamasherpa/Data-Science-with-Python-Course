## Lambda

```python
# Write a function to calculate the volume of  cuboid
def vol(l,b,h):
    v=l*b*h
    return v
```

```python
vol(21,65,23)
```

```python
v=lambda l,b,h : l*b*h
```

```python
v(10,21,31)
```

```python
# using lambda calculate the area of a triangle (1/2) *b* h
area = lambda b,h : (1/2) *b*h
```

```python
area(12,6)
```

## Map and Filters

```python
nums = [1,2,3,4,5]
```

```python
def sqr(x):
    return x**2
```

```python
result = map(sqr, nums)
result
```

```python
list(result)
```

```python
res=[]
for x in nums:
    res.append(x**2)
print(res)
```

```python
dumy=[10,12,13]
```

```python
sqr_d=list(map(sqr,dumy))
sqr_d
```

## Filter

```python
# filter(function, iterables)
numbers = list(range(2,99,3))
print(numbers, end=' ')
```

```python
def check(x):
    r = x % 2
    if r == 0:
        return x**2
```

```python
output= list(filter(check,numbers))
output
```

## List Comprehension

```python
given = list(range(1,11))
given
```

```python
sqrs=[]
for x in given:
    sqrs.append(x**2)
```

```python
sqrs
```

```python
output = [x**2 for x in given]
output
```

```python
# comprehension with condition
div6 = [x for x in range(1,101) if x%6==0]
div6
```

```python
div10= [x for x in range(1,201) if x%5==0 and x%7 == 0]
print(div10, end=' ')
```
